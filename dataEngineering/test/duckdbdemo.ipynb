##  by SELECT

```python
# 
conn.execute("SELECT * from weather").df()
```

```python
# 
conn.execute(f"""
  SELECT city, temp_lo, temp_hi, prcp, date
  FROM weather;
""").df()
```

```python
# 
conn.execute(f"""
SELECT city, (temp_hi + temp_lo) / 2 AS temp_avg, date
FROM weather;
""").df()
```

```python
#  by  (WHERE  )
conn.execute(f"""
SELECT *
FROM weather
WHERE city = 'San Francisco'
  AND prcp > 0.0;
""").df()
```

```python
#
conn.execute(f"""
SELECT *
FROM weather
ORDER BY city;
""").pl()
```

```python
conn.execute(f"""
SELECT *
FROM weather
ORDER BY city;
""").df()
```

```python
# by  (HAVING)
conn.execute(f"""
SELECT city, max(temp_lo)
FROM weather
GROUP BY city
HAVING max(temp_lo) < 40;
""").df()
```

```python
conn.execute(f"""
SELECT city, max(temp_lo)
FROM weather
WHERE city LIKE 'S%'
GROUP BY city
HAVING max(temp_lo) < 40;
""").df()
```
