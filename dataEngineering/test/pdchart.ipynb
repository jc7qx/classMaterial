## scatter plot

```python
import pandas as pd
df = pd.read_csv('E:/teaching/classMaterial/dataEngineering/data/titanic/train.csv')
```

```python
df.columns
```

```python
df.plot.scatter(x='Age', y='Fare', title='test', xlabel='age', ylabel='fare', grid=True)
```

## line plot

```python
df.plot.line(x='Fare', y='Age', marker="o", linestyle='-', title='test', xlabel='fare', ylabel='age', legend=False, grid=True)
```

## histogram

```python
df['Age'].plot.hist(title='titanic age dist', bins=15)
```

## box plot

```python
df.boxplot(column='Fare', by='Pclass', grid=True, vert=False)
```

## bar plot

```python
df['Sex'].value_counts().plot.bar(title='Employee Count by Department')
```

## area plot

```python
df.plot.area(x='AgeGroup', y='CumulativeSalaryByAgeGroup', title='Cumulative Salary Distribution Over Age Groups', xlabel='Age Group', ylabel='Cumulative Salary', legend=False, grid=True)
```

## pie chart

```python
df.groupby('Pclass')['Fare'].sum().plot.pie(title='Department-wise Salary Distribution', autopct='%1.1f%%')
```

```python
df['Pclass'].value_counts().plot.pie(autopct='%1.1f%%')
```
